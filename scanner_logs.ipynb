## Import package functions

```python
import matplotlib.pyplot as plt
```

```python
from scanner_logs import triolog, prismalog
```

## Trio respiration log

```python
t, data = triolog("log_trio.resp", doplot=True)
```

```python
# plot short segment
imin, imax = 70000, 80000
plt.figure(figsize=(20, 4)); plt.plot(t[imin:imax], data[imin:imax], '-k'); plt.show()
```
